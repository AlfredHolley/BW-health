```python
import sqlite3
import pandas as pd

# Connexion Ã  la base SQLite
conn = sqlite3.connect('database.sqlite')

# Lecture de la table patients dans un DataFrame pandas
df = pd.read_sql_query("SELECT code, startDate, active, isAdmin, language FROM patients", conn)

# Affichage du tableau
display(df)

# Fermeture de la connexion
conn.close()
```
